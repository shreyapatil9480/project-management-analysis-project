
# Project Management Analysis

This project provides a synthetic project management dataset and demonstrates exploratory data analysis (EDA) and predictive modeling techniques that a **Business Analyst**, **Program Manager** or **Data Analyst** might use to assess project performance.

## Dataset

The synthetic dataset includes information on 200 simulated projects with the following fields:

- `project_id` – Unique identifier for each project.
- `team_size` – Number of team members.
- `planned_duration_days` – Planned project duration (days).
- `actual_duration_days` – Actual project duration (days).
- `budget_k` – Planned budget (kilo dollars).
- `actual_cost_k` – Actual cost (kilo dollars).
- `complexity_score` – Complexity rating (1–10).
- `risk_score` – Risk rating (1–10).
- `total_tasks` – Number of tasks.
- `tasks_completed_mid` – Number of tasks completed halfway through the project.
- `schedule_variance_days` – Difference between actual and planned duration (days).
- `cost_variance_k` – Difference between actual and planned cost (kilo dollars).
- `project_success` – Binary indicator (1 = success, 0 = not success) based on schedule and cost performance plus risk rating.

The goal is to explore relationships between features and build models that predict project success and variances.  



```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error

# Set plotting styles
sns.set(style="whitegrid")
%matplotlib inline
```

```

# Load the synthetic dataset
file_path = '../synthetic_project_data.csv'
df = pd.read_csv(file_path)

# Display first few rows
df.head()
```

```

# Summary statistics
summary = df.describe()
summary
```

```

# Distribution plots
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
sns.histplot(df['schedule_variance_days'], bins=30, kde=True, color='steelblue')
plt.title('Schedule Variance Distribution (days)')
plt.xlabel('Schedule Variance (days)')

plt.subplot(1,2,2)
sns.histplot(df['cost_variance_k'], bins=30, kde=True, color='darkorange')
plt.title('Cost Variance Distribution (k$)')
plt.xlabel('Cost Variance (k$)')

plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(10,8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Heatmap')
plt.show()

# Scatter plot of complexity vs. schedule variance colored by success
plt.figure(figsize=(8,6))
sns.scatterplot(x='complexity_score', y='schedule_variance_days', hue='project_success', data=df, palette='Set1')
plt.title('Complexity vs Schedule Variance')
plt.xlabel('Complexity Score')
plt.ylabel('Schedule Variance (days)')
plt.show()
```

```

# Predictive modeling: classification of project success

# Feature selection
features = ['team_size', 'planned_duration_days', 'budget_k', 'complexity_score', 'risk_score',
            'total_tasks', 'tasks_completed_mid', 'schedule_variance_days', 'cost_variance_k']
X = df[features]
y = df['project_success']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Logistic Regression model
log_clf = LogisticRegression(max_iter=1000)
log_clf.fit(X_train, y_train)
y_pred_log = log_clf.predict(X_test)

print("
Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_log))
print("
Classification Report (Logistic Regression):
", classification_report(y_test, y_pred_log))

# Random Forest Classifier
rf_clf = RandomForestClassifier(n_estimators=200, random_state=42)
rf_clf.fit(X_train, y_train)
y_pred_rf = rf_clf.predict(X_test)

print("
Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("
Classification Report (Random Forest):
", classification_report(y_test, y_pred_rf))
```

```

# Predictive modeling: regression to predict cost variance

X_reg = df[['team_size', 'planned_duration_days', 'budget_k', 'complexity_score', 'risk_score', 'total_tasks', 'tasks_completed_mid']]
y_reg = df['cost_variance_k']

# Train-test split
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.3, random_state=42)

# Linear Regression model
lin_reg = LinearRegression()
lin_reg.fit(X_train_reg, y_train_reg)
y_pred_reg = lin_reg.predict(X_test_reg)

mse = mean_squared_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mse)

print("
Linear Regression RMSE:", rmse)
```
